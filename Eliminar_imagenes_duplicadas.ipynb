```
#contendrá las diferentes carpetas resultado de la búsqueda de imagenes
folders = [] 

#contendrá las carpetas con imágenes de cada compañia de alquiler
folders_ecooltra = []
folders_yego = []
folders_muving = []
folders_acciona = []
folders_blinkee = []
folders_motos = []


for r, d, f in os.walk(path):
  
    for folder in d:
        folders.append(os.path.join(r, folder))
        
        if 'ecooltra' in folder:
          folders_ecooltra.append(os.path.join(r, folder))
          
        elif 'muving' in folder:
          folders_muving.append(os.path.join(r, folder))
        
        elif 'yego' in folder:
          folders_yego.append(os.path.join(r, folder))
          
        elif 'acciona' in folder:
          folders_acciona.append(os.path.join(r, folder))
        
        elif 'Blinkee' in folder:
          folders_blinkee.append(os.path.join(r, folder))
        
        else:
          folders_motos.append(os.path.join(r, folder))
```

```
#definimos una función para listar todas las imagenes de una misma compañia de alquiler

def listarImagenes (carpetas, listaImagenes):
  
  for numfolders in range(len(carpetas)):
  
    for r, d, f in os.walk(carpetas[numfolders]):
        for file in f:
                listaImagenes.append(os.path.join(r, file))

  for f in listaImagenes:
      print(f)
  
  return True
```

```
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f): # @calcular_tiempo
   
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
  
```

```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```
@calcular_tiempo
def BuscarImagenesDuplicadas(image_files, basefolder_path):
  num_image2write=0
  save_baseimage = True
  num_imag_dup=0
  
  for numfiles in range(len(image_files)):
    file_base = image_files[0]
    print(file_base)
    image_files.remove(file_base) #eliminamos de la lista la imágen inicial que usamos como base de comparación para evitar comparaciones duplicadas en los siguientes loops
    baseImage = cv2.imread(file_base)
    
    for numfiles2 in range(len(image_files)): 
      compareImage = cv2.imread(image_files[numfiles2])

      #¿Mismo tamaño?
      if baseImage.shape == compareImage.shape: #Comparamos si coincide el tamaño de la imagen 

        # Si coincide el tamaño verificamos si tienen los mismos canales (colores)
        difference = cv2.subtract(baseImage, compareImage)
        b, g, r = cv2.split(difference)

        if cv2.countNonZero(b) == 0 and cv2.countNonZero(g) == 0 and cv2.countNonZero(r) == 0:
          num_imag_dup = num_imag_dup + 1
          save_baseimage = False
          break; #si encontramos una igual, salimos del for

        else: #Si no coincide en canales, guardamos la imagen, puesto que no es la misma. Guardamos la imagen base que estamabos comparando
          save_baseimage = True

      #Diferente tamaño
      else: #Si no coincide en tamaño, guardamos la imagen, puesto que no es la misma. Guardamos la imagen base que estamabos comparando
        save_baseimage = True

    if save_baseimage == True:
      cv2.imwrite(basefolder_path+"image"+str(num_image2write)+".jpg",baseImage)
      num_image2write = num_image2write +1
      save_baseimage = False

  print("Número de imagenes duplicadas en las búsquedas:")
  print(num_imag_dup)
  return num_imag_dup
```
