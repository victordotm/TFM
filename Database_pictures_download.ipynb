# Búsqueda y descarga de las imagenes
El siguiente código realiza una busqueda de imagenes en internet y descarga las imagenes encontradas.<br><br>
Para realizar la busqueda se basa en la API de Bing: *Bing Search APIs v7* con dicha funcionalidad: https://api.cognitive.microsoft.com/bing/v7.0/images <br><br>
El módulo de Python *search_bing_api.py* realiza las llamdas a la función correspondiente de la API de Bing, descargar las imagenes y las guarda. <br><br>

Buscamos imagenes de las siguientes compañias:


*   Ecooltra
*   Muving
*   Yego
*   Acciona
*   Motos en general 

<br>
Para cada compañia empleamos 3 diferentes argumentos de búsqueda, en una búsqueda independiente cada uno, con tal de conseguir un mayor número de resultados.<br>
Guardamos cada resultado de la búsqueda en una carpeta independiente.
